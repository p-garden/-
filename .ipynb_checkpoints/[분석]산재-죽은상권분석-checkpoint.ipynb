```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    
```

```python
# CSV 파일 불러오기
df = pd.read_csv("매장용빌딩+임대료·공실률+및+수익률.csv")

# 데이터프레임의 첫 5행 출력
print(df.head(5))
```

```python
# 열 이름 변경하여 사용하기 쉽게 변경
df.columns = ['Region_1', 'Region_2'] + [f"{year}_{col}" for year in range(2015, 2024) for col in ['rent', 'vacancy_rate', 'total_yield', 'income_yield', 'capital_yield']]
```

```python
# '소계' 행 제거
df = df[df['Region_2'] != '소계']

# 임대료, 공실률, 총 수익률에 대한 열만 필터링
filtered_data = df[['Region_1', 'Region_2'] + [f"{year}_{metric}" for year in range(2015, 2024) for metric in ['rent', 'vacancy_rate', 'total_yield']]]

# 숫자 값으로 변환
for col in filtered_data.columns[2:]:
    filtered_data[col] = pd.to_numeric(filtered_data[col], errors='coerce')
```

```python
# NaN 값 제거
filtered_data.dropna(inplace=True)

# 데이터프레임 출력
print(filtered_data.head())
```

```python
# 2015년부터 2023년까지 임대료, 공실률, 총 수익률의 변동률 계산

# Region_2 기준으로 그룹화하여 각 지역의 평균 변동률 계산
grouped_data = filtered_data.groupby('Region_2').mean(numeric_only=True)
```

```python
# 각 지표에 대한 변동률 계산
metrics = ['rent', 'vacancy_rate', 'total_yield']
changes_region2 = {}
for metric in metrics:
    changes_region2[f"{metric}_change"] = ((grouped_data[f'2023_{metric}'] - grouped_data[f'2015_{metric}']) / grouped_data[f'2015_{metric}']) * 100
```

```python
# 변동률 데이터를 지역 데이터와 결합
change_data_region2 = pd.DataFrame(changes_region2).reset_index()

# 무한대 값과 NaN 값 대체 및 제거
change_data_region2.replace([float('inf'), float('-inf')], pd.NA, inplace=True)
change_data_region2.dropna(inplace=True)

# 변동률 데이터프레임 출력
print(change_data_region2.head())
```

```python
# 시각화: 각 지표에 대한 변동률
fig, ax = plt.subplots(3, 1, figsize=(12, 18))

# 임대료 변동률
ax[0].bar(change_data_region2['Region_2'], change_data_region2['rent_change'], color='blue')
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Rent Change (%)')
ax[0].set_title('Rent Change by Region (2015-2023)')
ax[0].set_xticklabels(change_data_region2['Region_2'], rotation=45, ha='right', fontsize=10)

# 공실률 변동률
ax[1].bar(change_data_region2['Region_2'], change_data_region2['vacancy_rate_change'], color='red')
ax[1].set_xlabel('Region')
ax[1].set_ylabel('Vacancy Rate Change (%)')
ax[1].set_title('Vacancy Rate Change by Region (2015-2023)')
ax[1].set_xticklabels(change_data_region2['Region_2'], rotation=45, ha='right', fontsize=10)

# 수익률 변동률
ax[2].bar(change_data_region2['Region_2'], change_data_region2['total_yield_change'], color='green')
ax[2].set_xlabel('Region')
ax[2].set_ylabel('Total Yield Change (%)')
ax[2].set_title('Total Yield Change by Region (2015-2023)')
ax[2].set_xticklabels(change_data_region2['Region_2'], rotation=45, ha='right', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# 임대료가 가장 많이 하락한 지역 식별
declining_rent_region2 = change_data_region2.sort_values(by='rent_change').head(10)
print("임대료가 가장 많이 하락한 지역:")
print(declining_rent_region2)

# 공실률이 가장 많이 증가한 지역 식별
increasing_vacancy_region2 = change_data_region2.sort_values(by='vacancy_rate_change', ascending=False).head(10)
print("공실률이 가장 많이 증가한 지역:")
print(increasing_vacancy_region2)

# 수익률이 가장 많이 감소한 지역 식별
declining_yield_region2 = change_data_region2.sort_values(by='total_yield_change').head(10)
print("수익률이 가장 많이 감소한 지역:")
print(declining_yield_region2)
```

```python
# 임대료가 가장 많이 하락한 지역 식별
declining_rent_region2 = change_data_region2.sort_values(by='rent_change').head(10)
print("임대료가 가장 많이 하락한 지역:")
print(declining_rent_region2)
```

```python
# 시각화: 임대료가 가장 많이 하락한 지역
plt.figure(figsize=(12, 6))
plt.bar(declining_rent_region2['Region_2'], declining_rent_region2['rent_change'], color='blue')
plt.xlabel('Region')
plt.ylabel('Rent Change (%)')
plt.title('Top 10 임대료가 가장 많이 하락한 지역')
plt.xticks(ticks=range(len(declining_rent_region2['Region_2'])), labels=declining_rent_region2['Region_2'], rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# 공실률이 가장 많이 증가한 지역 식별
increasing_vacancy_region2 = change_data_region2.sort_values(by='vacancy_rate_change', ascending=False).head(10)
print("공실률이 가장 많이 증가한 지역:")
print(increasing_vacancy_region2)
```

```python
# 시각화: 공실률이 가장 많이 증가한 지역
plt.figure(figsize=(12, 6))
plt.bar(increasing_vacancy_region2['Region_2'], increasing_vacancy_region2['vacancy_rate_change'], color='red')
plt.xlabel('Region', fontproperties=fontprop)
plt.ylabel('Vacancy Rate Change (%)', fontproperties=fontprop)
plt.title('Top 10 Regions by Increase in Vacancy Rate', fontproperties=fontprop)
plt.xticks(ticks=range(len(increasing_vacancy_region2['Region_2'])), labels=increasing_vacancy_region2['Region_2'], rotation=45, ha='right', fontsize=10, fontproperties=fontprop)
plt.tight_layout()
plt.show()
```

```python
#수익률이 가장 많이 감소한 지역 식별
declining_yield_region2 = change_data_region2.sort_values(by='total_yield_change').head(10)
print("수익률이 가장 많이 감소한 지역:")
print(declining_yield_region2)
```

```python
# 종합 점수 계산
change_data_region2['total_score'] = (
    change_data_region2['rent_change'] +
    change_data_region2['vacancy_rate_change'] +
    change_data_region2['total_yield_change']
)

# 종합 점수로 상위 10개 지역 식별
top_declining_regions = change_data_region2.sort_values(by='total_score').head(10)
print("침체된 상권 (종합 점수 기준):")
print(top_declining_regions)

# 시각화: 종합 점수 기준 침체된 상권
plt.figure(figsize=(12, 6))
plt.bar(top_declining_regions['Region_2'], top_declining_regions['total_score'], color='purple')
plt.xlabel('Region')
plt.ylabel('Total Score')
plt.title('Top 10 Declining Regions by Total Score')
plt.xticks(ticks=range(len(top_declining_regions)), labels=top_declining_regions['Region_2'], rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
